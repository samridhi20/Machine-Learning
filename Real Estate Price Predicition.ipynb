# Treating : Outliers

```python
Q1 = df.Nearest_MRT_station.quantile(0.25)
Q3 = df.Nearest_MRT_station.quantile(0.75)

IQR = Q3 - Q1

df = df[(df.Nearest_MRT_station >= Q1 - 1.5 *IQR) & (df.Nearest_MRT_station <= Q3 + 1.5 *IQR)]


Q1 = df.Latitude.quantile(0.25)
Q3 = df.Latitude.quantile(0.75)

IQR = Q3 - Q1

df = df[(df.Latitude >= Q1 - 1.5 *IQR) & (df.Latitude <= Q3 + 1.5 *IQR)]

Q1 = df.Latitude.quantile(0.25)
Q3 = df.Logitude.quantile(0.75)

IQR = Q3 - Q1

df = df[(df.Logitude >= Q1 - 1.5 *IQR) & (df.Logitude <= Q3 + 1.5 *IQR)]

Q1 = df.Unit_area_price.quantile(0.25)
Q3 = df.Unit_area_price.quantile(0.75)

IQR = Q3 - Q1

df = df[(df.Unit_area_price >= Q1 - 1.5 *IQR) & (df.Unit_area_price <= Q3 + 1.5 *IQR)]
```

# Train-Test-Split

```python
from sklearn.model_selection import train_test_split
from mlxtend.evaluate import bias_variance_decomp
from sklearn.metrics import mean_absolute_error as mae
```

```python
Y = df.pop('Unit_area_price')
x = df
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,Y,test_size=0.3,random_state=33)
```

```python
print("X Train",x_train.shape)
print("X Test",x_test.shape)
print("Y Train",y_train.shape)
print("Y Test",y_test.shape)
```

# Model Building 

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression(normalize=True)
```

```python
LR.fit(x_train,y_train)
```

```python
# Model Coeffiencent
print(LR.coef_) 
# Model intercept
print(LR.intercept_)
```

```python
y_train_predicted = LR.predict(x_train) # X train
y_test_predict = LR.predict(x_test) # X test
```

```python
k = mae(y_train_predicted,y_train)
print("Training Mean Absolute Error",k)
p = mae(y_test_predict,y_test)
print("Test Mean Absolute Error",p)
```

```python
# Model training score

LR.score(x_train,y_train)
```

```python
# Model testing score

LR.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
from math import sqrt
```

```python
# Checking MSE

MSE = mean_squared_error(y_test,y_test_predict)
print(MSE)
```

```python
# checking RMSE score

RMSE = sqrt(MSE)
print('RMSE',RMSE)

#checking MAPE

MAPE = mean_squared_error(y_train,y_train_predicted)
print(MAPE)
```

```python
# Checking R2 Score

r_square = r2_score(y_test,y_test_predict)
print(r_square)
```
